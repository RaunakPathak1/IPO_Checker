```python
dhrp_doc_collection = chroma_client.get_collection(COLLECTION_NAME)
```

```python
# Full doucment embeddings retrieval

full_result = dhrp_doc_collection.get(include=['embeddings', 'documents', 'metadatas'])
dhrp_vectors = np.array(full_result['embeddings'])
dhrp_documents = full_result['documents']
dhrp_metadatas = full_result['metadatas']
dhrp_doc_types = [metadata['source'] for metadata in dhrp_metadatas]
full_colors = [['blue', 'red'][['DRHP_KSH International IPO.PDF', 'Insulin'].index(t)] for t in dhrp_doc_types]
```

```python
print(dhrp_doc_types)
```

```python
# Full doucment embeddings retrieval visualization
# We humans find it easier to visalize things in 2D!
# Reduce the dimensionality of the vectors to 2D using t-SNE
# (t-distributed stochastic neighbor embedding)

tsne = TSNE(n_components=2, random_state=42)
reduced_vectors = tsne.fit_transform(dhrp_vectors)

# Create the 2D scatter plot
fig = go.Figure(data=[go.Scatter(
    x=reduced_vectors[:, 0],
    y=reduced_vectors[:, 1],
    mode='markers',
    marker=dict(size=5, color=full_colors, opacity=0.8),
    text=[f"Type: {t}<br>Text: {d[:100]}..." for t, d in zip(dhrp_doc_types, dhrp_documents)],
    hoverinfo='text'
)])

fig.update_layout(title='2D Chroma Vector Store Visualization',
    scene=dict(xaxis_title='x',yaxis_title='y'),
    width=800,
    height=600,
    margin=dict(r=20, b=10, l=10, t=40)
)

fig.show()
```
